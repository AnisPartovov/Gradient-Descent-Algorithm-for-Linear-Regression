```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def gradient_descent(X, Y, learning_rate=0.01, num_iterations=1000):
  
    m = 0
    b = 0
    N = float(len(Y))
    
    for i in range(num_iterations):
        Y_pred = m*X + b
        error = Y_pred - Y
        m -= (learning_rate/N) * np.dot(error, X)
        b -= (learning_rate/N) * np.sum(error)
    
    return b, m


def mse(X, Y, m, b):
    Y_pred = m*X + b
    return np.mean((Y_pred - Y)**2)


def generate_data(slope, intercept, num_samples=100, noise_std=1):
    np.random.seed(0)
    X = np.linspace(0, 10, num_samples)
    Y = slope*X + intercept + np.random.randn(num_samples)*noise_std
    return X, Y

slope = 2
intercept = 1
X, Y = generate_data(slope, intercept)

print("True intercept:", intercept)
print("True slope:", slope)
```

```python

learning_rates = [0.001, 0.01]

fig, axs = plt.subplots(1, 2, figsize=(16, 6))
for i, lr in enumerate(learning_rates):

    b, m = gradient_descent(X, Y, learning_rate=lr)
    print(b, m)
    
    Y_pred = m*X + b
    axs[i].scatter(X, Y, color='k', label='Data')
    axs[i].plot(X, Y_pred, color="r", label=f"LR={lr}")
    axs[i].set_xlabel('X', fontsize=12)
    axs[i].set_ylabel('Y', fontsize=12)
    axs[i].tick_params(axis='both', labelsize=10)
    axs[i].set_title(f'LR={lr}', fontsize=14)

plt.suptitle('Linear Regression using Gradient Descent', fontsize=16)
plt.show()
```
